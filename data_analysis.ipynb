```python
import pandas as pd

df = pd.read_csv('apartments_for_rent_classified_10K.csv', encoding="Windows-1252", sep=';', header=0)
df.head()
```

```python
print(df.shape)
```

```python

# Display basic statistics for the split dataframe
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df = df.dropna(how='any')
df = df[~df.isin([float('nan')]).any(axis=1)]
df.head()
```

```python
df.isnull().sum()
```

```python
print(df.shape)
```

```python
numeric_cols = ['bathrooms', 'bedrooms', 'price', 'square_feet', 'latitude', 'longitude', 'time']
numeric_df = df[numeric_cols]
# Converting categorical columns to numeric where possible
numeric_df['number_of_amenities'] = df['amenities'].apply(lambda x: len(x.split(',')) if isinstance(x, str) else 0)
numeric_df.to_csv('apartments_for_rent_10k_numeric.csv', index=False)
```
